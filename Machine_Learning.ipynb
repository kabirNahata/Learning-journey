```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df

```

```python
ldf=df[['area','price']]
```

```python
ldf
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(ldf.area,ldf.price,color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(ldf[['area']],ldf.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# y=m*x+b
167.30954677*3300+76692.38187078107
```

```python
%matplotlib inline
plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(ldf.area,ldf.price,color='red',marker='+')
plt.plot(ldf.area,reg.predict(ldf[['area']]),color='blue')
```

# Multiple Variables

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
```

```python
df
```

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms =df.bedrooms.fillna(median_bedrooms)
```

```python
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```
